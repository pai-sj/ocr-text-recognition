# 3. 모델 구성하기



```python
from models.layers import ConvFeatureExtractor, Map2Sequence 
from models.layers import BLSTMEncoder, CTCDecoder, DotAttention

import tensorflow as tf
from tensorflow.keras import backend as K
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.layers import GRU
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import TimeDistributed
from tensorflow.keras.layers import concatenate
```

```python
onset_classes = train_gen.onset_classes
nucleus_classes = train_gen.nucleus_classes
coda_classes = train_gen.coda_classes
```

```python
K.clear_session()

height = 36
onset_classes = train_gen.onset_classes
nucleus_classes = train_gen.nucleus_classes
coda_classes = train_gen.coda_classes

n_conv = 16 # the number of Convolution filter
n_state = 128 # the number of GRU units
n_embed = 16 # the size of embedding vector
n_clf = 128 # the number of units in classifier Dense layer

# Image Encoder
inputs = Input(shape=(height, None, 1),name='images')
transposed = K.permute_dimensions(inputs, (0, 2, 1, 3))
conv_maps = ConvFeatureExtractor(n_conv,
                                 name='feature_extractor')(transposed)
feature_seqs = Map2Sequence(name='map_to_sequence')(conv_maps)
dense_seqs = Dense(n_state, activation='tanh')(feature_seqs)

# Embedding Layer
decoder_onset_inputs = Input(shape=(None,), name='decoder_onset_inputs')
decoder_nucleus_inputs = Input(shape=(None,), name='decoder_nucleus_inputs')
decoder_coda_inputs = Input(shape=(None,), name='decoder_coda_inputs')

onset_embedding_layer = Embedding(onset_classes+1, n_embed)
nucleus_embedding_layer = Embedding(nucleus_classes+1, n_embed)
coda_embedding_layer = Embedding(coda_classes+1, n_embed)

onset_embedding = onset_embedding_layer(decoder_onset_inputs)
nucleus_embedding = nucleus_embedding_layer(decoder_nucleus_inputs)
coda_embedding = coda_embedding_layer(decoder_coda_inputs)

concat_embedding = concatenate([onset_embedding, 
                                nucleus_embedding, 
                                coda_embedding],
                               axis=-1,
                               name='concat_embedding')

# Text Decoder
decoder_state_inputs = Input(shape=(n_state,), name='decoder_state')
gru_layer = GRU(n_state, 
                name='decoder_gru', 
                return_sequences=True)
states_decoder = gru_layer(concat_embedding,
                           initial_state=decoder_state_inputs)

# Attention Layer
dotattend = DotAttention()
context, attention = dotattend([dense_seqs, states_decoder])

# Classifier Layer
onset_clf1 = Dense(n_clf, activation='relu')
onset_clf2 = Dense(onset_classes+1, name='onset_seqs',
                   activation='softmax')

nucleus_clf1 = Dense(n_clf, activation='relu')
nucleus_clf2 = Dense(nucleus_classes+1, name='nucleus_seqs',
                     activation='softmax')

coda_clf1 = Dense(n_clf, activation='relu')
coda_clf2 = Dense(coda_classes+1, name='coda_seqs',
                  activation='softmax')

concat_output = concatenate([context, states_decoder],
                            axis=-1, name='concat_output')

onset_predictions = onset_clf2(onset_clf1(concat_output))
nucleus_predictions = nucleus_clf2(nucleus_clf1(concat_output))
coda_predictions = coda_clf2(coda_clf1(concat_output))
```

## (8) 추론 모델과 학습 모델 구성하기

```python
# For training
trainer = Model([inputs, 
                 decoder_onset_inputs,
                 decoder_nucleus_inputs,
                 decoder_coda_inputs,
                 decoder_state_inputs], 
                 [onset_predictions,
                  nucleus_predictions,
                  coda_predictions], name='trainer')

# # For Inference
# # - (1) Encoder
# encoder = Model(inputs, states_encoder, 
#                 name='encoder')

# # - (2) Decoder
# states_encoder_input = Input((None,n_lstm*2), 
#                              name='states_encoder_input')

# context, attention = dotattend([states_encoder_input, states_decoder])
# concat_output = concatenate([context, states_decoder], axis=-1, 
#                             name='concat_output')
# fc_outputs = TimeDistributed(clf1_layer)(concat_output)
# predictions = TimeDistributed(clf2_layer, name='output_seqs')(fc_outputs)

# decoder = Model([states_encoder_input, decoder_inputs, decoder_state_inputs], 
#                 [states_decoder, predictions], name='decoder')
```
